# Walk-Forward Backtest: Top-1 Momentum (NautilusTrader)

This notebook uses `momentum_walkforward_core.py` to evaluate multiple indices and parameter sets
with a walk-forward procedure, and visualize average performance per parameter set for each index.


```python
# If running in a fresh environment, uncomment to install dependencies
# %pip install yfinance pandas beautifulsoup4 lxml html5lib nautilus-trader matplotlib

import os, sys
# Make module importable
sys.path.append(os.path.abspath('stock_enhanced'))
from momentum_walkforward_core import run_walkforward
```

```python
# Global data window for downloads
GLOBAL_START = '2022-01-01'
GLOBAL_END   = '2025-07-31'

# Walk-forward configuration
TEST_WINDOW_DAYS = 120   # test window length
STEP_DAYS        = 120    # step between window starts

# Strategy parameter grid to test
PARAM_GRID = [
    { 'LOOKBACK': 150, 'ROC': 150, 'REB_DAYS': 14, 'POS_SIZE': 0.95 },
    { 'LOOKBACK': 200, 'ROC': 200, 'REB_DAYS': 14, 'POS_SIZE': 0.95 },
    { 'LOOKBACK': 250, 'ROC': 250, 'REB_DAYS': 14, 'POS_SIZE': 0.95 },
    { 'LOOKBACK': 200, 'ROC': 150, 'REB_DAYS': 7,  'POS_SIZE': 0.95 },
    { 'LOOKBACK': 200, 'ROC': 250, 'REB_DAYS': 21, 'POS_SIZE': 0.95 },
]

# Indices to evaluate
CANDIDATE_INDICES = ['SP500','SP500_IT','NASDAQ100','DAX40','FTSE100','CAC40','NIFTY50','KOSPI200']

# Initial capital and fixed per-trade fee (engine may handle fees internally)
INITIAL_CAPITAL = 10000000
TX_COST        = 7.0
```

```python
# Run the walk-forward evaluation
results = run_walkforward(
    global_start=GLOBAL_START,
    global_end=GLOBAL_END,
    param_grid=PARAM_GRID,
    candidate_indices=CANDIDATE_INDICES,
    test_window_days=TEST_WINDOW_DAYS,
    step_days=STEP_DAYS,
    initial_capital=INITIAL_CAPITAL,
    tx_cost=TX_COST,
    max_workers=8 ,
    run_in_parallel=True,
    verbose=True, 
)
```

```python
# Inspect aggregated results for a specific index, e.g., NASDAQ100
results.get('SP500')
```

```python
  import matplotlib.pyplot as plt                                                                                     
  import seaborn as sns                                                                                               
  import pandas as pd                                                                                                 
  import numpy as np                                                                                                  
                                                                                                                      
  # Assuming your results are stored in the 'results' variable                                                        
  # If you have results for multiple indices, this will analyze all of them                                           
                                                                                                                      
  def analyze_walkforward_results(results):                                                                           
      """                                                                                                             
      Comprehensive analysis and visualization of walk-forward results                                                
      """                                                                                                             
                                                                                                                      
      # Combine all indices into one DataFrame for comparison                                                         
      all_results = []                                                                                                
      for index_name, df in results.items():                                                                          
          df_copy = df.copy()                                                                                         
          df_copy['index'] = index_name                                                                               
          all_results.append(df_copy)                                                                                 
                                                                                                                      
      if not all_results:                                                                                             
          print("No results to analyze!")                                                                             
          return                                                                                                      
                                                                                                                      
      combined_df = pd.concat(all_results, ignore_index=True)                                                         
                                                                                                                      
      # Set up the plotting style                                                                                     
      plt.style.use('default')                                                                                        
      sns.set_palette("husl")                                                                                         
                                                                                                                      
      # Create a comprehensive dashboard                                                                              
      fig = plt.figure(figsize=(20, 16))                                                                              
                                                                                                                      
      # 1. Overall Performance Comparison                                                                             
      plt.subplot(3, 3, 1)                                                                                            
      if 'index' in combined_df.columns and len(combined_df['index'].unique()) > 1:                                   
          sns.boxplot(data=combined_df, x='index', y='ret_pct')                                                       
          plt.title('Return % Distribution by Index')                                                                 
          plt.xticks(rotation=45)                                                                                     
      else:                                                                                                           
          plt.hist(combined_df['ret_pct'], bins=20, alpha=0.7, edgecolor='black')                                     
          plt.title('Return % Distribution')                                                                          
          plt.xlabel('Return %')                                                                                      
          plt.ylabel('Frequency')                                                                                     
                                                                                                                      
      # 2. Parameter Impact - Lookback Period                                                                         
      plt.subplot(3, 3, 2)                                                                                            
      lookback_perf = combined_df.groupby('LOOKBACK')['ret_pct'].mean().sort_values(ascending=False)                  
      lookback_perf.plot(kind='bar', color='skyblue', edgecolor='black')                                              
      plt.title('Average Return % by Lookback Period')                                                                
      plt.xlabel('Lookback Days')                                                                                     
      plt.ylabel('Average Return %')                                                                                  
      plt.xticks(rotation=0)                                                                                          
                                                                                                                      
      # 3. Parameter Impact - ROC Period                                                                              
      plt.subplot(3, 3, 3)                                                                                            
      roc_perf = combined_df.groupby('ROC')['ret_pct'].mean().sort_values(ascending=False)                            
      roc_perf.plot(kind='bar', color='lightcoral', edgecolor='black')                                                
      plt.title('Average Return % by ROC Period')                                                                     
      plt.xlabel('ROC Days')                                                                                          
      plt.ylabel('Average Return %')                                                                                  
      plt.xticks(rotation=0)                                                                                          
                                                                                                                      
      # 4. Parameter Impact - Rebalancing Frequency                                                                   
      plt.subplot(3, 3, 4)                                                                                            
      reb_perf = combined_df.groupby('REB_DAYS')['ret_pct'].mean().sort_values(ascending=False)                       
      reb_perf.plot(kind='bar', color='lightgreen', edgecolor='black')                                                
      plt.title('Average Return % by Rebalancing Frequency')                                                          
      plt.xlabel('Rebalancing Days')                                                                                  
      plt.ylabel('Average Return %')                                                                                  
      plt.xticks(rotation=0)                                                                                          
                                                                                                                      
      # 5. Heatmap of Parameter Combinations                                                                          
      plt.subplot(3, 3, 5)                                                                                            
      pivot_table = combined_df.pivot_table(values='ret_pct',                                                         
                                           index='LOOKBACK',                                                          
                                           columns='ROC',                                                             
                                           aggfunc='mean')                                                            
      sns.heatmap(pivot_table, annot=True, fmt='.2f', cmap='RdYlGn', center=0)                                        
      plt.title('Return % Heatmap: Lookback vs ROC')                                                                  
                                                                                                                      
      # 6. Risk-Return Scatter                                                                                        
      plt.subplot(3, 3, 6)                                                                                            
      if len(combined_df) > 1:                                                                                        
          risk_return = combined_df.groupby(['LOOKBACK', 'ROC', 'REB_DAYS']).agg({                                    
              'ret_pct': ['mean', 'std']                                                                              
          }).round(3)                                                                                                 
          risk_return.columns = ['avg_return', 'volatility']                                                          
          risk_return = risk_return.reset_index()                                                                     
                                                                                                                      
          scatter = plt.scatter(risk_return['volatility'], risk_return['avg_return'],                                 
                              s=100, alpha=0.7, c=risk_return['avg_return'], cmap='RdYlGn')                           
          plt.xlabel('Volatility (Std Dev of Returns)')                                                               
          plt.ylabel('Average Return %')                                                                              
          plt.title('Risk-Return Profile')                                                                            
          plt.colorbar(scatter, label='Average Return %')                                                             
                                                                                                                      
      # 7. Top 10 Parameter Combinations                                                                              
      plt.subplot(3, 3, 7)                                                                                            
      top_10 = combined_df.nlargest(10, 'ret_pct')[['LOOKBACK', 'ROC', 'REB_DAYS', 'ret_pct']]                        
      top_10['param_combo'] = top_10.apply(lambda x: f"L{int(x.LOOKBACK)}-R{int(x.ROC)}-D{int(x.REB_DAYS)}", axis=1)  
                                                                                                                      
      plt.barh(range(len(top_10)), top_10['ret_pct'], color='gold', edgecolor='black')                                
      plt.yticks(range(len(top_10)), top_10['param_combo'])                                                           
      plt.xlabel('Return %')                                                                                          
      plt.title('Top 10 Parameter Combinations')                                                                      
      plt.gca().invert_yaxis()                                                                                        
                                                                                                                      
      # 8. Performance Distribution                                                                                   
      plt.subplot(3, 3, 8)                                                                                            
      plt.hist(combined_df['ret_pct'], bins=20, alpha=0.7, color='purple', edgecolor='black')                         
      plt.axvline(combined_df['ret_pct'].mean(), color='red', linestyle='--',                                         
                  label=f'Mean: {combined_df["ret_pct"].mean():.2f}%')                                                
      plt.axvline(combined_df['ret_pct'].median(), color='orange', linestyle='--',                                    
                  label=f'Median: {combined_df["ret_pct"].median():.2f}%')                                            
      plt.xlabel('Return %')                                                                                          
      plt.ylabel('Frequency')                                                                                         
      plt.title('Return Distribution')                                                                                
      plt.legend()                                                                                                    
                                                                                                                      
      # 9. Summary Statistics Table (text)                                                                            
      plt.subplot(3, 3, 9)                                                                                            
      plt.axis('off')                                                                                                 
                                                                                                                      
      stats_text = f"""                                                                                               
      PERFORMANCE SUMMARY                                                                                             
                                                                                                                      
      Best Return: {combined_df['ret_pct'].max():.2f}%                                                                
      Worst Return: {combined_df['ret_pct'].min():.2f}%                                                               
      Average Return: {combined_df['ret_pct'].mean():.2f}%                                                            
      Median Return: {combined_df['ret_pct'].median():.2f}%                                                           
                                                                                                                      
      Standard Deviation: {combined_df['ret_pct'].std():.2f}%                                                         
                                                                                                                      
      Positive Returns: {(combined_df['ret_pct'] > 0).sum()}/{len(combined_df)}                                       
      Win Rate: {(combined_df['ret_pct'] > 0).mean()*100:.1f}%                                                        
                                                                                                                      
      Best Lookback: {lookback_perf.index[0]} days                                                                    
      Best ROC: {roc_perf.index[0]} days                                                                              
      Best Rebalancing: {reb_perf.index[0]} days                                                                      
      """                                                                                                             
                                                                                                                      
      plt.text(0.1, 0.9, stats_text, transform=plt.gca().transAxes,                                                   
               fontsize=11, verticalalignment='top', fontfamily='monospace',                                          
               bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))                                         
                                                                                                                      
      plt.tight_layout()                                                                                              
      plt.show()                                                                                                      
                                                                                                                      
      # Detailed rankings table                                                                                       
      print("\n" + "="*80)                                                                                            
      print("DETAILED PARAMETER RANKING")                                                                             
      print("="*80)                                                                                                   
                                                                                                                      
      detailed_ranking = combined_df.sort_values('ret_pct', ascending=False).copy()                                   
      detailed_ranking['rank'] = range(1, len(detailed_ranking) + 1)                                                  
      detailed_ranking['net_pnl_formatted'] = detailed_ranking['net_pnl'].apply(lambda x: f"${x:,.0f}")               
                                                                                                                      
      display_cols = ['rank', 'LOOKBACK', 'ROC', 'REB_DAYS', 'POS_SIZE', 'ret_pct', 'net_pnl_formatted']              
      if 'index' in detailed_ranking.columns:                                                                         
          display_cols.insert(1, 'index')                                                                             
                                                                                                                      
      print(detailed_ranking[display_cols].to_string(index=False))                                                    
                                                                                                                      
      return combined_df                                                                                              
                                                                                                                      
  # Run the analysis                                                                                                  
  combined_results = analyze_walkforward_results(results)                                                             
                                                                                                                      
  # Additional correlation analysis                                                                                   
  print("\n" + "="*50)                                                                                                
  print("PARAMETER CORRELATION ANALYSIS")                                                                             
  print("="*50)                                                                                                       
                                                                                                                      
  correlation_matrix = combined_results[['LOOKBACK', 'ROC', 'REB_DAYS', 'POS_SIZE', 'ret_pct']].corr()                
  print("\nCorrelation with Return %:")                                                                               
  return_correlations = correlation_matrix['ret_pct'].sort_values(ascending=False)                                    
  for param, corr in return_correlations.items():                                                                     
      if param != 'ret_pct':                                                                                          
          print(f"{param:12}: {corr:6.3f}")                                                                           
                                                                                                                      
  # Show parameter value counts                                                                                       
  print("\n" + "="*50)                                                                                                
  print("PARAMETER FREQUENCY ANALYSIS")                                                                               
  print("="*50)                                                                                                       
                                                                                                                      
  for param in ['LOOKBACK', 'ROC', 'REB_DAYS']:                                                                       
      print(f"\n{param} Performance:")                                                                                
      param_stats = combined_results.groupby(param)['ret_pct'].agg(['count', 'mean', 'std', 'min', 'max']).round(2)   
      print(param_stats.to_string())  
```
